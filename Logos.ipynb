```python
from bs4 import BeautifulSoup
import pandas as pd
import os
from requests_html import HTMLSession
import requests
```

```python
data=pd.read_csv("~/Downloads/E0.csv")
```

```python
listofteams= data['HomeTeam'].unique()
```

```python
def scrape_premier_league_logos():
    url = "https://www.premierleague.com/clubs?se=274"
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Assuming logos are inside a specific class
        logo_elements = soup.find_all('div', class_='clubBadgeContainer')

        # Create a list to store logo URLs
        logos = []

        for logo_element in logo_elements:
            # Assuming logos are stored as background images in the style attribute
             img_element = logo_element.find('img')
             if img_element:
                logo_url = img_element['src']
                logos.append(logo_url)

        return logos
    else:
        print(f"Failed to fetch page. Status code: {response.status_code}")
        return None


```

```python
# Example usage
premier_league_logos = scrape_premier_league_logos()
if premier_league_logos:
    print("Premier League Logos:")
    for logo in premier_league_logos:
        print(logo)
```
